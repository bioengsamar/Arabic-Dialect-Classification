```python
from twarc import Twarc
import pandas as pd
from tqdm import tqdm
import os

def initiate_csv(file_name):
    file_name = pd.DataFrame([])
    return file_name

def download_QADI(consumer_key, consumer_secret, access_token, access_token_secret, ids_file, label, file_name):
  t = Twarc(consumer_key, consumer_secret, access_token, access_token_secret)
  file_ = initiate_csv(file_name)

  for tweet in tqdm(t.hydrate(open(ids_file))):
    file_ = file_.append(pd.DataFrame({'text': tweet["full_text"], 'dialect': label}, index=[0]), ignore_index=True)
  file_.to_csv('/content/drive/MyDrive/task_aim/data/data_text_'+label+'.csv', index=False)
  return file_
```

# Example of how to get the text (dialect) from the IDs files
